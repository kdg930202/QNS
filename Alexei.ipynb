# Single
$$
  E_\text{loc}(\sigma) = \sum_\eta \langle\sigma|H|\eta\rangle \frac{\psi(\eta)}{\psi(\sigma)} = \sum_{\eta\,\, \text{s.t.} \,\,\langle\sigma|H|\eta\rangle \neq 0} \langle\sigma|H|\eta\rangle \exp[\log\psi(\eta) - \log\psi(\sigma)] 
$$


# Double
$$
  E_\text{loc}(\sigma)_{n,n+1} = \sum_\eta \langle\sigma|H|\eta\rangle \frac{\psi_{n+1}(\eta)}{\psi_{n}(\sigma)} = \sum_{\eta\,\, \text{s.t.} \,\,\langle\sigma|H|\eta\rangle \neq 0} \langle\sigma|H|\eta\rangle \exp[\log\psi_{n+1}(\eta) - \log\psi_n(\sigma)] 
$$


```python
def to_array(model, parameters):
    # begin by generating all configurations in the hilbert space.
    # all_States returns a batch of configurations that is (hi.n_states, N) large.
    all_configurations = hi.all_states()

    # now evaluate the model, and convert to a normalised wavefunction.
    logpsi = model.apply(parameters, all_configurations)
    psi = jnp.exp(logpsi)
    psi = psi / jnp.linalg.norm(psi)
    return psi
```

```python
# todelete
def local_energies_single(model, parameters, hamiltonian_jax, sigma):
    eta, H_sigmaeta = hamiltonian_jax.get_conn_padded(sigma)
    
    logpsi_sigma = model.apply(parameters, sigma)
    logpsi_eta = model.apply(parameters, eta)
    logpsi_sigma = jnp.expand_dims(logpsi_sigma, -1) 
    
    res = jnp.sum(H_sigmaeta * jnp.exp(logpsi_eta - logpsi_sigma), axis=-1)
    
    return res

def local_energies_double(model, parameter1, parameter2, hamiltonian_jax, sigma):
    #parameter2 : (n+1) psi
    #parameter1 : (n) psi
    
    eta, H_sigmaeta = hamiltonian_jax.get_conn_padded(sigma)
    
    logpsi_sigma = model.apply(parameter2, sigma)
    logpsi_eta = model.apply(parameter1, eta)
    logpsi_sigma = jnp.expand_dims(logpsi_sigma, -1) 
    
    res = jnp.sum(H_sigmaeta * jnp.exp(logpsi_eta - logpsi_sigma), axis=-1)
    
    return res
```

## `jax.vjp`

**A bit about math**

Mathematically, suppose we have a function $f: \mathbb{R}^n \rightarrow \mathbb{R}^m$, the Jacobian matrix of $f$ at a particular point $x$, denoted $J(x) \in \mathbb{R}^{m \times n}$, is a matrix:
$$J(x) = 
\left(\begin{matrix} 
\frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \cdots & \frac{\partial f_m}{\partial x_n}
\end{matrix} \right)$$

You can think of it as a linear map $J(x): \mathbb{R}^n \rightarrow \mathbb{R}^m$ which maps $v$ to $J(x)v$.

What vector-Jacobian product does is to compute $vJ(x)$ or $J(x)^\top v$. `jax.vjp` is the api to compute the vector-Jacobian product in JAX with two arguments:
- first argument: a callable function $f$
- second argument: primal value at which point the Jacobian is evaluated (Should be either a tuple or a list of arguments)

It returns both $f(x)$ and a linear map $J(x)^\top: \mathbb{R}^m \rightarrow \mathbb{R}^n$ which map $v$ to $J^\top v$.

```python
def estimate_gradient(model, para_before, para_after, ha_Et_jax, ha_jax_squared, sigma):
    # reshape the samples to a vector of samples with no extra batch dimensions

    sigma = sigma.reshape(-1, sigma.shape[-1])
    
    E_loc_squared = local_energies_single(model, para_after, ha_jax_squared, sigma)
    E_loc_before = local_energies_single(model, para_before, ha_Et_jax, sigma)
    E_loc_after = local_energies_single(model, para_after, ha_Et_jax, sigma)
    E_loc_double = local_energies_double(model, para_before, para_after, ha_Et_jax, sigma)
    
    # compute the energy as well
    E_squared_average = jnp.mean(E_loc_squared)
    E_double_average = jnp.mean(E_loc_double)
    E_after_average = jnp.mean(E_loc_after)

    logpsi_sigma_fun = lambda pars : model.apply(pars, sigma)
    
    
    #With respect to theta_(n+1) ????????????????????????????
    _, vjpfun_before = jax.vjp(logpsi_sigma_fun, para_before)
    _, vjpfun_after = jax.vjp(logpsi_sigma_fun, para_after)
    
    E_grad_1 = vjpfun_after((E_loc_squared - E_squared_average)/E_loc_after.size)
    
    E_grad_2 = vjpfun_before(-2*(E_loc_after - E_loc_double)/E_loc_after.size)
    E_grad_3 = vjpfun_after(-2*(E_loc_before - E_loc_double)/E_loc_after.size)

    E_grad = (E_grad_1 + E_grad_2 + E_grad_3)
    return E_grad[0]
```

```python
def variance(H, psi):
    return np.conj(psi)@H@H@psi - (np.conj(psi)@H@psi)**2
```

```python
num_chain = 20
n_iters = 1000
l_rate = 0.001
# vs_init = vs.parameters
para_before = ma.init(jax.random.PRNGKey(0), np.ones((hi.size, )))
para_after = ma.init(jax.random.PRNGKey(0), np.ones((hi.size, )))
sampler_state = sa.init_state(ma, para_before, seed=1)
result = []
for i in tqdm(range(n_iters)):
    # sample
    
    sampler_state = sa.reset(ma, para_before, state=sampler_state)
    samples, sampler_state = sa.sample(ma, para_before, state=sampler_state, chain_length=num_chain)
    
    # compute energy and gradient
#     E, E_grad = estimate_energy_and_gradient(model, parameters, hamiltonian_jax, samples)
    E_grad = estimate_gradient(ma, para_before, para_after, ha_Et_jax, ha_squared_jax, samples)
    # update parameters. Try using a learning rate of 0.01
    para_before = jax.tree_map(lambda x,y: x-l_rate*y, para_after, E_grad)
    para_after = para_before
    psi = to_array(ma,para_after)
#     result = variance(ha.to_dense(), psi)
    result.append(variance(ha.to_dense(), psi))
# print(result)    
```
